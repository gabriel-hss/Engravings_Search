```python
import pandas as pd
```

```python
engravings = pd.read_excel("dataset_engravings_24classes.xlsx")
```

```python
def search_engraving(dataframe: pd.DataFrame, engraving1: str, engraving2: str, status: str, acessory: str):
    engraving1_cond = dataframe['Mandatory'].str.contains(engraving1) | dataframe['Secondary'].str.contains(engraving1)
    engraving2_cond = dataframe['Mandatory'].str.contains(engraving2) | dataframe['Secondary'].str.contains(engraving2)

    filtered_dataframe = dataframe[engraving1_cond & engraving2_cond]

    if acessory == "all":
        status_cond = (
            filtered_dataframe['Necklace'].str.contains(status)
            | filtered_dataframe['Earring'].str.contains(status)
            | filtered_dataframe['Ring'].str.contains(status)
        )
    else:
        status_cond = filtered_dataframe[acessory].str.contains(status)
        
    return filtered_dataframe[status_cond]
```

```python
search_engraving(engravings, "Mayhem", "Adrenaline", status="Spec", acessory="All")
```
